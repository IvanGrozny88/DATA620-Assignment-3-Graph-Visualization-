Laod the libraries


```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
from pylab import rcParams
```

```python
# Set the parameters
%matplotlib inline
rcParams['figure.figsize'] = 5, 4
```

Read the facebook data

```python
# Read 500 rows
dt = pd.read_table("C:/Users/Ivant/Desktop/facebook_combined (2).txt/facebook_combined.txt", sep = " ", names = ['source','target'])
```

```python
dt = dt[300:400]
```

```python
dt.head()
```

```python
dt.shape
```

Make a graph object

```python
G = nx.from_pandas_edgelist(dt)
```

```python
print( nx.info(G))
```

The graph has 94 nodes, 100 edges

Diameter

```python
nx.algorithms.diameter(G)
```

The diameter of the graph is 8.

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = (10, 6)
nx.draw(G, node_color='green', with_labels=True)
```
